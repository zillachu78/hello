```python
# Name of the Experiment: Write the Python code to perform Data Cleaning techniques using a suitable Dataset.

# Step 1: Import Necessary Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Step 2: Load the Dataset
df = pd.read_csv(r"house perice.csv")  # Make sure the filename is correct

# Step 3: Initial Data Exploration
print(df.head(10))       # Display first 10 rows
print(df.info())         # Data types and missing values
print(df.describe())     # Summary statistics

# Step 4: Data Cleaning Techniques

## 4.1 Handling Missing Values
# df.drop(columns=['Dimensions', 'Plot Area'], inplace=True)  # Drop irrelevant columns
df['SalePrice'].fillna(df['SalePrice'].median(), inplace=True)  # Fill missing price with median

# Fill missing categorical values with mode
categorical_cols = df.select_dtypes(include='object').columns
for col in categorical_cols:
    df[col].fillna(df[col].mode()[0], inplace=True)

print("Missing values after filling:\n", df.isnull().sum())

## 4.2 Handling Duplicate Values
duplicate_count = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicate_count}")
df.drop_duplicates(inplace=True)
print(f"Data shape after removing duplicates: {df.shape}")

## 4.3 Dropping Remaining Null Values
print("Null values before dropping:\n", df.isnull().sum())
df.dropna(inplace=True)
print(f"Shape after dropping nulls: {df.shape}")
print("Null values after dropping:\n", df.isnull().sum())

## 4.4 Handling Outliers in Price
Q1 = df['SalePrice'].quantile(0.25)
Q3 = df['SalePrice'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df = df[(df['SalePrice'] >= lower_bound) & (df['SalePrice'] <= upper_bound)]
print(f"Data shape after removing outliers: {df.shape}")
```
